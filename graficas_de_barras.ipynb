````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import userdata
````

````python
ruta = userdata.get('ruta')

df = pd.read_csv(ruta + 'melbourne_housing-clean.csv', index_col=0)

df.head()
````

````python
from google.colab import drive
drive.mount('/content/drive')
````

````python
rooms_count = df['rooms'].value_counts()
````

````python
# Crear un gráfico de barras con Seaborn
ax = sns.barplot(x=rooms_count.index, y=rooms_count, order=rooms_count.index)

plt.figure(figsize=(6, 10))

# Configurar el título y las etiquetas
ax.set_title('Conteo de casas por cantidad de habitaciones')
ax.set(xlabel='Cantidad de Habitaciones', ylabel='Cantidad de casa')

# Mostrar el gráfico
plt.show()
````

````python
type_counts = df['type'].value_counts()
type_counts = type_counts * 100 / type_counts.sum()
````

````python
# Crear un gráfico de barras con Seaborn
ax = sns.barplot(x=type_counts.index, y=type_counts)

plt.figure(figsize=(6, 10))

# Configurar el título y las etiquetas
ax.set_title('Procentage por tipo de categorias')
ax.set(xlabel='Categorias', ylabel='Porcentaje %')

# Mostrar el gráfico
plt.show()
````

````python
council_counts = df['council_area'].value_counts()
````

````python
# Crear un gráfico de barras con Seaborn
ax = sns.barplot(x=council_counts, y=council_counts.index, orient='h')

plt.figure(figsize=(6, 10))

# Configurar el título y las etiquetas
ax.set_title('Distribución por consejos')
ax.set(xlabel='Frecuencia', ylabel='Consejo')

# Mostrar el gráfico
plt.show()
````

````python
regionname_type_rooms = pd.crosstab(df['region_name'], [df['type'], df['rooms']])
regionname_type_rooms
````

````python
regionname_type_rooms_stacked = regionname_type_rooms.stack(level=0, future_stack=True)
regionname_type_rooms_stacked
````

````python
# Crear un gráfico de barras con Seaborn
ax = sns.barplot(x=regionname_type_rooms_stacked.loc[('Eastern Metropolitan', 'h')].index,
            y=regionname_type_rooms_stacked.loc[('Eastern Metropolitan', 'h')])

plt.figure(figsize=(6, 10))

# Configurar el título y las etiquetas
ax.set_title('Conteo de casas por cantidad de habitaciones de tipo "h" en Eastern Metropolitan')
ax.set(xlabel='Cantidad de Habitaciones', ylabel='Cantidad de casas')

# Mostrar el gráfico
plt.show()
````
